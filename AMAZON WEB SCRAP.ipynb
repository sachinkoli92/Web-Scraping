```python
import requests
from bs4 import BeautifulSoup
```

```python
#Requesting webpage server to get the source code using requests
page2=requests.get('https://www.amazon.in/s?i=electronics&bbn=1389432031&rh=n%3A1389432031%2Cp_36%3A-2000000&dc&page=2&qid=1612522831&rnid=1318502031&ref=sr_pg_2')
```

```python
#parsing the code using html.parser
page_2=BeautifulSoup(page2.content,'html.parser')
page_2
```

```python
#Extracting tags containing mobile names
mn=page_2.find_all('span', class_="a-size-base-plus a-color-base a-text-normal")
mn
```

```python
#Empty lists to store the records
Mobile=[]
Price=[]
Rating=[]
Image_Url=[]
```

```python
#using for loop to fetch the records from the tag
for i in mn:
    Mobile.append(i.text)
Mobile
```

```python
Mobile[0]
```

```python
#Extracting tags of price
price=page_2.find_all('span', class_="a-price-whole")
price
```

```python
#using for loop fetching the records of price
for i in price:
    Price.append(i.text)
Price
```

```python
#Extracting tags of rating
rate=page_2.find_all('span', class_="a-icon-alt")
rate
```

```python
#using for loop to fetch the ratings
for i in rate:
    Rating.append(i.text)
Rating
```

```python
#extracting codes for url
url=page_2.find_all('div', class_="a-section aok-relative s-image-square-aspect")
url
```

```python
link=[]
import re
for i in url:
    for j in i.find_all('img',{'src':re.compile('.jpg')}):
        print(j)
        link.append(j)
```

```python
link
```

```python
URL=[]
for p in link:
    print(p['src'])
    URL.append(p['src'])
```

```python
URL
```

```python
#creating a dataframe to store the records
import pandas as pd
info=pd.DataFrame({})
info['Mobile']=Mobile
info['Price']=Price
info['Rating']=Rating[4:28]
info['URL']=URL
```
